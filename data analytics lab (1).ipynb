```python
def assess_water_quality(pH, DO, TDS, Nitrate, Fecal_Coliform,Turbidity):
    issues = []
    recommendations = []

    # pH Analysis
    if pH < 6.5:
        issues.append(" Water is too acidic.")
        recommendations.append(" Add alkaline substances (e.g., lime) to neutralize pH.")
    elif pH > 8.5:
        issues.append(" Water is too alkaline.")
        recommendations.append(" Introduce CO2 injection or mild acids to balance pH.")
    
    # Dissolved Oxygen (DO)
    if DO < 4:
        issues.append(" Oxygen levels critically low.")
        recommendations.append(" Increase aeration using fountains, aerators, or artificial oxygenation.")

    # Total Dissolved Solids (TDS)
    if TDS > 1000:
        issues.append(" High TDS levels detected.")
        recommendations.append(" Use RO (Reverse Osmosis) filtration or distillation for purification.")

    # Nitrate Levels
    if Nitrate > 50:
        issues.append(" Excessive nitrate levels detected.")
        recommendations.append(" Reduce fertilizer runoff, implement buffer zones, and improve agricultural practices.")

    # Fecal Coliform Contamination
    if Fecal_Coliform > 100:
        issues.append(" High bacterial contamination detected.")
        recommendations.append(" Improve sewage treatment, boil water before consumption, and maintain sanitation.")
        
    # Turbidity
    if Turbidity > 10:
        issues.append(" High Turbidity detected.")
        recommendations.append("Use filtration (sand, activated carbon) to reduce turbidity.")
    if Turbidity > 50:
        issues.append(" Critical Warning unsafe for drinking.")
        recommendations.append("Apply coagulation, sedimentation, and advanced filtration.")

    # Overall Water Quality Status
    if not issues:
        quality_status = " Water is **SAFE** for drinking."
    elif len(issues) <= 2:
        quality_status = " Water quality is **MODERATE**  Needs minor improvements."
    else:
        quality_status = " Water quality is **POOR**  Immediate action required!"

    return quality_status, issues, recommendations

# User Input
print("\n **Water Quality Assessment Tool** ")
pH = float(input("Enter pH Level (Recommended: 6.5 - 8.5): "))
DO = float(input("Enter Dissolved Oxygen (mg/L) (Recommended: >4): "))
TDS = int(input("Enter Total Dissolved Solids (mg/L) (Recommended: <1000): "))
Nitrate = int(input("Enter Nitrate Level (mg/L) (Recommended: <50): "))
Fecal_Coliform = int(input("Enter Fecal Coliform (MPN/100mL) (Recommended: <100): "))
Turbidity = int(input("Enter Turbidity (NTU) (Recommended: <15): "))

# Get Water Quality Report
status, detected_issues, actions = assess_water_quality(pH, DO, TDS, Nitrate, Fecal_Coliform,Turbidity)

# Display Report
print("\n **Water Quality Report** ")
print(status)

if detected_issues:
    print("\n **Issues Detected:**")
    for issue in detected_issues:
        print(f"  - {issue}")

if actions:
    print("\n **Recommended Actions:**")
    for action in actions:
        print(f"  - {action}")

print("\n Follow these recommendations to ensure safe and clean water! ")
```

```python
bad_thresholds = {
    'TDS (mg/L)': 500,
    'Nitrate (mg/L)': 10,
    'DO (mg/L)': 4,  # Lower DO is bad
    'Fecal Coliform (MPN/100mL)': 5,
    'Turbidity (NTU)': 20
}

# Count number of violations per region
df['Bad_Water_Score'] = (
    (df['TDS (mg/L)'] > bad_thresholds['TDS (mg/L)']).astype(int) +
    (df['Nitrate (mg/L)'] > bad_thresholds['Nitrate (mg/L)']).astype(int) +
    (df['DO (mg/L)'] < bad_thresholds['DO (mg/L)']).astype(int) +  # Low DO is bad
    (df['Fecal Coliform (MPN/100mL)'] > bad_thresholds['Fecal Coliform (MPN/100mL)']).astype(int) +
    (df['Turbidity (NTU)'] > bad_thresholds['Turbidity (NTU)']).astype(int)
)

# Sort regions by highest contamination score
df_sorted = df.sort_values(by='Bad_Water_Score', ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Bad_Water_Score', data=df_sorted, palette="Reds")

plt.xlabel("Region")
plt.ylabel("Number of Poor Quality Indicators")
plt.title("Regions with Poor Water Quality")
plt.ylim(0, max(df['Bad_Water_Score']) + 1)

# Show plot
plt.show()
```
